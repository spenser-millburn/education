## Question 1
Analytically show that the product of a complex number ùëß and its conjugate is a
real number |ùëß|2, i.e., the square of the magnitude of the complex number.

### Proof:

1. The product of $z$ and $\overline{z}$ is :
   $$z\overline{z} = (a + bi)(a - bi)$$

2. distributive property :
   $$z\overline{z} = a^2 - (bi)^2$$

3. $i^2 = -1$, so substitute $i^2$ with $-1$ and simplify:
   $$z\overline{z} = a^2 + b^2$$

4. The square of the magnitude of $z$ is:
   $$|z|^2 = (\sqrt{a^2 + b^2})^2 = a^2 + b^2$$

$$z\overline{z} = |z|^2$$

## Question 7 
Determine an expression for a sinusoid that
oscillates 15 times per second, that has a value
of -1 at t = 0, and whose peak amplitude is 3.
Use MATLAB to plot the signal over 0 ‚â§ t ‚â§ 1

```python
import numpy as np
import matplotlib.pyplot as plt

A = 3  # Amplitude
f = 15  # Frequency in Hz
value_at_t0 = -1

# Calculate phi 
phi = np.arcsin(value_at_t0 / A)

# Time vector from 0 to 1 second
t = np.linspace(0, 1, 1000)  # 1000 points for smooth plotting

# Sinusoid expression
y = A * np.sin(2 * np.pi * f * t + phi)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, y, label='Sinusoid')
plt.title('Sinusoid Oscillating 15 Times per Second')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

## Question 8

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the time range
t = np.linspace(0, 1, 500)  # from 0 to 1 with 500 points

# Define the signals
xa = np.exp(-t)
xb = np.sin(2 * np.pi * (5 * t))
xc = np.exp(-t) * np.sin(2 * np.pi * (5 * t))

# Plot the signals
plt.figure(figsize=(10, 6))
plt.plot(t, xa, label=r'$x_a(t) = e^{-t}$')
plt.plot(t, xb, label=r'$x_b(t) = \sin(2\pi 5t)$')
plt.plot(t, xc, label=r'$x_c(t) = e^{-t} \sin(2\pi 5t)$')

plt.title('Sketch of Signals')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the coefficient matrix and the constant vector
A = Matrix([
    [1, 1, 1],
    [1, 2, 3],
    [1, -1, 0]
])
b = Matrix([1, 3, -3])

# Compute determinants needed for Cramer's Rule
det_A = A.det()

# For x1
A_x1 = A.copy()
A_x1[:, 0] = b
det_A_x1 = A_x1.det()

# For x2
A_x2 = A.copy()
A_x2[:, 1] = b
det_A_x2 = A_x2.det()

# For x3
A_x3 = A.copy()
A_x3[:, 2] = b
det_A_x3 = A_x3.det()

# Calculate x1, x2, x3 using Cramer's Rule
x1 = det_A_x1 / det_A
x2 = det_A_x2 / det_A
x3 = det_A_x3 / det_A

x1, x2, x3
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function x(t) as defined in the image provided by the user
def x(t):
    if t >= -1 and t < 0:
        return t + 1
    elif t >= 0 and t <= 1:
        return 1 - t
    else:
        return 0

# Generating time points
t = np.linspace(-2, 2, 1000)
# Applying the function to time points
x_t = np.array([x(ti) for ti in t])

# Plotting the function x(t)
plt.figure(figsize=(10, 4))
plt.plot(t, x_t, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Signal x(t)')
plt.grid(True)
plt.legend()
plt.show()
```

\begin{align*}
\text{Original signal:} \quad & x(t) \\
\text{Amplitude scaling:} \quad & v(t) = 3 \cdot x(t) \\
\text{Time scaling:} \quad & v(t) =  (-\frac{1}{2} \cdot t) \\
\text{Time shifting:} \quad & v(t) = (t + 1) \\
\end{align*}


```python
# Define the transformed signal v(t) = 3x(-(1/2)(t + 1))
def v(t):
    # Apply the transformation to the time variable before passing it to the x(t) function
    transformed_t = -0.5 * (t + 1)
    return 3 * x(transformed_t)

# Apply the function to each time point using list comprehension
v_values = [v(ti) for ti in t]

# Plot the graph
plt.plot(t, v_values, label='v(t) = 3x(-(1/2)(t + 1))')
plt.title('Graph of v(t)')
plt.xlabel('t')
plt.ylabel('v(t)')
plt.grid(True)
plt.legend()
plt.show()
```

# Question 9 
Problem 1.11-1 of the textbook.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time interval and number of points
t = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Define the original function x(t) with the unit step function behavior included
def u(t):
    return np.where(t >= np.pi, 1, 0)

def x(t):
    return (2 - t * np.cos(2*np.pi*t)) * u(t)

# Calculate the odd portion xo(t)
def xo(t):
    return 0.5 * (x(t) - x(-t))

# Plot xo(t)
plt.figure(figsize=(10, 6))
plt.plot(t, xo(t), label='$x_o(t)$')
plt.title('Odd Portion $x_o(t)$ of the Function $x(t)$')
plt.xlabel('$t$')
plt.ylabel('$x_o(t)$')
plt.grid(True)
plt.legend()
plt.show()
```

# Question 10

```python
# Define the time interval and spacing
t = np.arange(-2, 2, 0.002)  

# Define the original function x(t) with the updated parameters
def u(t):
    return np.where(t >= 0.5, 1, 0)

def x(t):
    return (2 - t/2 * np.cos(4*np.pi*t)) * u(t)

# Calculate the even portion xe(t)
def xe(t):
    return 0.5 * (x(t) + x(-t))

# Plot xe(t)
plt.figure(figsize=(10, 6))
plt.plot(t, xe(t), label='$x_e(t)$')
plt.title('Even Portion $x_e(t)$ of the Function $x(t)$')
plt.xlabel('$t$')
plt.ylabel('$x_e(t)$')
plt.grid(True)
plt.legend()
plt.show()
```

## [78] Multiplying matrices by vectors
- Similar to matrix multiplication, but using dot product


**To multiply a matrice by a vector**
- Always produces another row or column vector of same dimensions
    - multiply columns of matrice by row vector **OR**
    - multiply rows of matrice by column vector 

$
AB =
\left[ 
    \begin{array}[]\
        1 & 0 & -2 & 3\\
        0 & 4 & 3 & -1\\
    \end{array}
\right]
\cdot
\left[ 
    \begin{array}[]\
        1 \\ 0 \\ -2 \\ 0\
    \end{array}
\right]
$

=
$
\left[ 
    \begin{array}[]\
        1(1)+0(0)-2(-2)+3(0)\\
        0(1)+4(0)+3(-2)-1(0)\\
    \end{array}
\right]
$

=
$
\left[ 
    \begin{array}[]\
        5\\
        -6\\
    \end{array}
\right]
$

## [81] The null space 
### Formulae
**[81 definition of the null space]** : $A \vec{x} = \vec{0}$
- Testing if a vector is in the null space : **dot product after RREF is 0**

$
RREF(
\left[ 
    \begin{array}[]\
        3 & 1 & 8\\
        1 & 1 & 1\\
    \end{array}
\right]
\cdot
\left[ 
    \begin{array}[]\
        2 \\ -3\\ 1\
    \end{array}
\right]
) =
\left[ 
    \begin{array}[]\
        0 \\ 0\\ 0\
    \end{array}
\right]
$                     -- Therefore vector (2,-3,1) is in the null space 

- To find all vectors in the null space
    - Plug the matrix into equation [81] and dot the matrix with the vector 
    - Write as an augmented matrix
    - reduce to RREF
    - pull the solutions out of the augmented matrix
    - *See below example*
    - The span of the vectors obtained from this method contain all vectors in the null space of the matrix

```python
#Testing for null pace ( matrix dotted with vector)
import numpy as np
from sympy import * 
A = np.matrix([[-3, 1,9], [1,1,1]])
M = Matrix(np.dot(A,np.matrix([[2],[-3],[1]])))
M_rref = M.rref()  
display(M_rref[0])
```

**RREF Matrix**

$
\left[ 
    \begin{array}[]\
        1 & 0 & -6\\
        0 & 1 & -4\\
        0 & 0 & 0\\
    \end{array}
\right]
\left[ 
    \begin{array}[]\
        x_1 \\ x_2 \\ x_3\\
    \end{array}
\right]
$

**Convert to standard notation, and solve for pivot variables x_1,x_2**
- Coefficients of free variables are placed in vector of last step
$x_1 - 6x_3 = 0 \rightarrow x_1 = 6x_3$

$x_2 -4x_3 = 0 \rightarrow x_2 = 4x_3$

$x_3-x_3 = 0 \rightarrow x_3 = x_3$ 

$
N(M) = SPAN
\left(
\left[ 
    \begin{array}[]\
        6 \\ 4\\ 1\\
    \end{array}
\right]
\right)
$

```python
#[Quiz 33] Finding all vectors in the null space of a matrix with sympy
import numpy as np
from sympy import *
M = Matrix([[1,2,-3,5],[2,4,-6,10],[-3,-6,9,-15],[4,1,-12,6]])
display("Matrix M",M)
M_rref = M.rref()
display("RREF of M",M_rref[0])
display("Pivot columns",M_rref[1])

M_nullspace = M.nullspace()
for i, nspace in enumerate(M_nullspace):
    display(f"nullspacex_{i}",nspace)
    
```

## [85] The column space
- All combinations of the columns in a 
    - AKA the span of columns in A
- Always a subspace    
- Linearly dependent
- Does not form a basis
### Formulae

### Solving for the general solution of $A \vec{x} = b$
$A\vec{x} = \vec{0} \rightarrow$ Complementary Solution $\rightarrow A\vec{x_n} =\vec{0}$\
$A\vec{x} = \vec{b} \rightarrow$ Particular Solution $\rightarrow A\vec{x_p} =\vec{b}$
- Add these equations to get general solution
$A \vec{x} = b \rightarrow A(\vec{x_n} + \vec{x_p}) = \vec{b}$

#### Process for solving a general solution to  $A \vec{x} = b$

$
A = 
\left[ 
    \begin{array}{}\
        3 & -6 & 6  \\
        -3 & 7 & -9 \\
        -6 & 8 & 0 \\
    \end{array}
\right]
$

$\vec{b} = 
\left[ 
    \begin{array}{}\
        1  \\
        -1 \\
        -2 \\
    \end{array}
\right]
$

#### 1. Augment the matrix for 0 and $\vec{b}$ as two separate matrices or in parallel

$
A = 
\left[ 
    \begin{array}{ccc|c}\
        3 & -6 & 6 & 0\\
        -3 & 7 & -9 & 0\\
        -6 & 8 & 0 & 0\\
    \end{array}
\right]
$

$
B = 
\left[ 
    \begin{array}{ccc|c}\
        3 & -6 & 6 & b_1\\
        -3 & 7 & -9 & b_2 \\
        -6 & 8 & 0 & b_3\\
    \end{array}
\right]
$

##### 2. reduce to row-echelon form


$
A = 
\left[ 
    \begin{array}{ccc|c}\
        1 & 0 & -4 & 0\\
        0 & 1 & -3 & 0\\
        0 & 0 & 0 & 0\\
    \end{array}
\right]
$

$
B = 
\left[ 
    \begin{array}{ccc|c}\
        1 & 0 & -4 & 2(b_1+b_2)+\frac{1}{3}b_1 \rightarrow \frac{7}{3}b_1 + 2b_2\\
        0 & 1 & -3 & b_1 + b_2\\
        0 & 0 &  0 & 6b_1 + 4b_2 +b_3\\
    \end{array}
\right]
$

#### 3. Capture the null space vector(s) from $A$ and plug in $\vec{b}$ values to $B$ and capture vectors 

    - Converting A to a system of equations

Pivot variable : $x_1 = 4x_3 = 4$

Pivot variable : $x_2 = 3x_3 = 3$  

Free variable : $x_3 = x_3 = 1$

    - Capturing vector from $A$ system of equations
$\vec{x_n} = c1
\left[ 
    \begin{array}{}\
         4\\
         3\\
         1\\
    \end{array}
\right]
$

    - Plugging values into $B$
$    
\frac{7}{3}b_1 + 2b_2 = \frac{7}{3}(1) + 2(-1) = \frac{1}{3}\\
b_1 + b_2 = 1+(-1) = 0 \\
6b_1 + 4b_2 +b_3 = 6(1) + 4(-1)+(-2) = 0\\
$

    - Converting $B$ to a system of equations

Pivot variable : $x_1 = 4x_3 = \frac{1}{3}$

Pivot variable : $x_2 = 3x_3 = 0$  

Free variable : $x_3 = x_3 = 0$

    - Capturing vector from $B$ equations
$\vec{x_p} = 
\left[ 
    \begin{array}{}\
         \frac{1}{3}\\
         0\\
         0\\
    \end{array}
\right]
$

#### 4. general solution is acquired by adding the vectors

$
\vec{x} = 
\left[ 
    \begin{array}{}\
         \frac{1}{3}\\
         0\\
         0\\
    \end{array}
\right]
+ c_1 \left[ 
    \begin{array}{}\
         4\\
         3\\
         1\\
    \end{array}
\right]
$

## [88] Dimensionality : Nullity And Rank
- **Nullity** is given by the number of basis vectors required to define the null space of a matrix
- **Rank** is given by the number of pivot columns in the RREF of a matrix

```python
#[88] Finding the rank of a matrix
import numpy as np
from sympy import *
M = Matrix([[1,2,-3,5],[2,4,-6,10],[-3,-6,9,-15],[4,1,-12,6]])
display("M", M)
M_rref = M.rref()

M_nullspace = M.nullspace()
display("RREF",M_rref[0])

print("\nRank\n\n",(M_rref[1]))
```

```python
#[88] Finding the rank  and nullity of a matrix
import numpy as np
from sympy import *
M = Matrix([[1,3,-2,-1,0],[2,5,-4,-7,3],[1,4,-3,5,4],[1,2,-2,-6,3]])
display("M", M)

M_nullspace = M.nullspace()
for i, nullspace in enumerate(M_nullspace):
    display(f"Nullspace {i}",nullspace)
    
display("NULLITY : ", len (M_nullspace))
display("RREF",M_rref[0])

print("\nRank\n\n",len(M_rref[1]))
```

## Formulae
**[62] The square of the length of the vector is equal to the vector dotted with itself**\
$||\vec{a}||^2 = a_1^2+a_2^2 = \vec{a} \cdot \vec{a} = a_1 a_1 + a_2 a_2$

**[64] Cauchy-Scharz Inequality**\
$|\vec{u}\cdot\vec{v}| \le ||\vec{u}||\cdot||\vec(v)||$\
    - Dependent (colinear) case: $\vec{u}\cdot\vec{v} \lt ||\vec{u}||\cdot||\vec(v)||$\
    - Inependent case: $\vec{u}\cdot\vec{v} = ||\vec{u}||\cdot||\vec(v)||$ 
    
**[65] Vector Triangle Inequality**\
$||\vec{u}+\vec{v}|| \le ||\vec{u}||+||\vec{v}||$\
    - Dependent (colinear) case: $||\vec{u}\cdot\vec{v} || = ||\vec{u}||+||\vec{v}|| $\
    - Inependent case: $||\vec{u}\cdot\vec{v}|| \lt 0 $ 
    
**[67] Angle Between Two Vectors**\
$\vec{u}\cdot\vec{v} = ||\vec{u}||||\vec{v}||\cos(\theta)$\
$\theta = acos(\frac{\vec{u}\cdot\vec{v}}{ ||\vec{u}||||\vec{v}||})$

## [61] Dot Products
### Perpendicular and orthogonal Vectors
- When the dot product of two non-zero vectors is 0, the vectors must be perpendicular
- When the dot product of n<2 non-zero vectors is 0, the vectors must be orthogonal 

### **[62] Properties of the Dot Product**
- Commutative - $\vec{a} \cdot \vec{b} = a_1 b_1 + a_2 b_2 =\vec{b} \cdot \vec{a} = b_1 a_1 + b_2 a_2 $
- Distributive - $\vec{a} \cdot ( \vec{b}+\vec{c} ) = \vec{a} \cdot \vec{b} + \vec{a} \cdot \vec{c}$
- Associative - $(3\vec{a} \cdot \vec{b}) = 3(\vec{a} \cdot \vec{b})$

```python
#[64]Cauchy-Schwarz
import numpy as np
import math

u_set = [[3,2],[6,5],[1,1],[8,-7]]
v_set = [[-12,-8],[4,0],[2,2],[-16,14]]

def cauchy_schwarz(u,v,independent=True):
    result_dot_product= abs(np.dot(u, v))
    result_length_product = round(np.linalg.norm(u)*np.linalg.norm(v))
    if(independent):
        result = (f"{result_dot_product} < {result_length_product} : ", result_dot_product<result_length_product)
    else:
        result = (f"{result_dot_product} == {result_length_product} : ", result_dot_product==result_length_product)
    return result 

for i,u in enumerate(u_set):
    print(cauchy_schwarz(u,v_set[i]))
    
```

```python
#[65]Vector Triangle
import numpy as np
import math

u_set = np.matrix([[-9,2],[3,-1],[5,-3],[-9,3]])
v_set = np.matrix([[54,-12],[8,-3],[0,6],[-72,24]])

def vector_triangle(u,v,independent=True):
    result_vecsum_length = round(np.linalg.norm(u + v),3)
    result_length_sum = round(np.linalg.norm(u)+np.linalg.norm(v),3)
    if(independent):
        result = (f"{result_vecsum_length} < {result_length_sum} : ", result_vecsum_length<result_length_sum)
    else:
        result = (f"{result_vecsum_length} == {result_length_sum} : ", result_vecsum_length==result_length_sum)
    return result 

for i,u in enumerate(u_set):
    print(vector_triangle(u,v_set[i]))
```

```python
#[67]Angle between two vectors
import numpy as np
import math

def unit_vector(vector):
    """ Returns the unit vector of the vector.  """
    return vector / np.linalg.norm(vector)

def angle_between(v1, v2):
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))

u = np.array([1,-3,1])
v = np.array([0,6,-2])

math.degrees(angle_between(u,v))
```


## [69] Normal Vectors and Planes
- **Planes**
    - A plane is defined as the set of *all vectors that are orthogonal to one given normal vector*
        - D shifts the plane, x y z define the orientation
    - To **find the equation of a plane from two given points*
        - plug the positional points into eq[69], Plug the normal vector in as A B C and simplify
- **Normal Vectors**
    - The normal vector is the the vector that is orthogonal to all of the vectors in a plane
    - Every plane has two normal vectors (positive and negative)
    - To **find the equation of a plane from a given normal vector*
        - pull out the constants from eq[69] as $\vec{n} = (3,-2,-5)$ 

```python
#[70] Finding the equation of a plane from three points
def equation_plane(x1, y1, z1, x2, y2, z2, x3, y3, z3):
     
    a1 = x2 - x1
    b1 = y2 - y1
    c1 = z2 - z1
    a2 = x3 - x1
    b2 = y3 - y1
    c2 = z3 - z1
    a = b1 * c2 - b2 * c1
    b = a2 * c1 - a1 * c2
    c = a1 * b2 - b1 * a2
    d = (- a * x1 - b * y1 - c * z1)
    print ("equation of plane is ", a, "x +", b, "y +", c, "z +", d, "= 0.")
 
# Driver Code

x1,y1,z1,x2,y2,z2,x3,y3,z3 = (-1), 2, 1, 0, (-3), 2, 1, 1, (-4)

equation_plane(x1, y1, z1, x2, y2, z2, x3, y3, z3)
```

```python
#[70] finding the equation of a plane from normal vector and plotting
import numpy
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plane parameters
n_1=2
n_2=0
n_3=1

x_0=0
y_0=0
z_0=10

# Create a grid of x and y

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xg, yg = np.meshgrid(x, y)

# Compute z = f(x,y)

z = -n_1/n_3*(xg-x_0)-n_2/n_3*(yg-y_0)+z_0 

ax.plot_wireframe(xg, yg, z, rstride=10, cstride=10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plane: z=f(x,y)')
plt.show()
```

## [72] Cross Products
- The result of a cross product is a resultant vector that is orthogonal to the vectors.

- *Process for taking the cross product of a determinate*
    1. Determinate form and Signs
        - $\vec{a}\times\vec{b} = 
\begin{vmatrix}
i & j & k \\ 
a_1 & a_2 & a_3 \\ 
b_1 & b_2 & b_3 \\ 
\end{vmatrix}\
\begin{vmatrix}
+ & - & + \\ 
- & + & -\\ 
+ & - & +\\ 
\end{vmatrix}
$
    2. *Reduce to 2x2 Determinates, assign signs by referencing the sign matrix above*
        - $
\vec{a}\times\vec{b} = 
i
\begin{vmatrix}
a_2 & a_3  \\ 
b_2 & b_3 \\ 
\end{vmatrix}\
-j
\begin{vmatrix}
a_1 & a_3  \\ 
b_1 & b_3 \\ 
\end{vmatrix}\
+k
\begin{vmatrix}
a_1 & a_2  \\ 
b_1 & b_2 \\ 
\end{vmatrix}\
$
    3. *Cross Multiply and subtract from left to right*
        - $ i(a_2 b_3 - a_3 b_2) -j(a_1 b_3 - a_3 b_1) + k(a_1b_2 - a_2 b_1)$

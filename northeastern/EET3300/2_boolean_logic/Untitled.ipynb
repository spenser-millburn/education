```python
import ttg
def tt(args_list,exprs = None):
    df =  (ttg.Truths(args_list,exprs).as_pandas())[::-1]
    df = df.reset_index(drop = True)
    return df
    
```

```python
#2
tt(['A', 'B', 'C'], 
   ['(A and C) or (~A and ~C) or (B and C) or (~A and B)',                    
    '(~A or B or C) and (~A or ~B or C) and (A or B or ~C)'
   ])
```

```python
#4
tt(['A', 'B', 'C'], 
   ['B and (~A or C) or (A and C)',
   ])
```

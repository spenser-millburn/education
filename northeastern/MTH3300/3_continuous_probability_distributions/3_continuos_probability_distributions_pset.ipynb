### Problem 14: Elongation of Steel Plates

Given the probability density function (pdf) for elongation $ f(x) = \frac{x}{250} $ for $ 20 < x < 30 $, we solve several sub-problems.

#### a. Proportion of steel plates with elongations greater than 25%

$ P(X > 25) = \int_{25}^{30} \frac{x}{250} \, dx $

Calculating this integral:

$ P(X > 25) = \left[ \frac{x^2}{500} \right]_{25}^{30} = \frac{30^2}{500} - \frac{25^2}{500} = \frac{900}{500} - \frac{625}{500} = \frac{275}{500} = 0.55 $

Proportion: $0.55$

#### b. Mean elongation

$ \mu = \int_{20}^{30} x \cdot \frac{x}{250} \, dx $

Calculating this integral:

$ \mu = \int_{20}^{30} \frac{x^2}{250} \, dx = \left[ \frac{x^3}{750} \right]_{20}^{30} = \frac{30^3}{750} - \frac{20^3}{750} = \frac{27000}{750} - \frac{8000}{750} = \frac{19000}{750} = 25.33 $

Mean elongation: $25.33$

#### c. Variance of the elongations

First, find $ E[X^2] $:

$ E[X^2] = \int_{20}^{30} x^2 \frac{x}{250} \, dx = \int_{20}^{30} \frac{x^3}{250} \, dx = \left[ \frac{x^4}{1000} \right]_{20}^{30} = \frac{30^4}{1000} - \frac{20^4}{1000} = 650 $

Now, calculate the variance:

$ \sigma^2 = E[X^2] - \mu^2 = 650 - (25.33)^2 = 650 - 641.11 = 8.89 $

Variance: $8.89$

#### d. Standard deviation of the elongations

$ \sigma = \sqrt{\sigma^2} = \sqrt{8.89} = 2.98 $

Standard deviation: $2.98$

#### e. Cumulative distribution function (CDF)

$ F(x) = \int_{20}^{x} \frac{t}{250} \, dt = \left[ \frac{t^2}{500} \right]_{20}^{x} = \frac{x^2}{500} - 0.8 $

CDF: 
$ F(x) = \begin{cases} 
0 & x \leq 20 \\
\frac{x^2}{500} - 0.8 & 20 < x < 30 \\
1 & x \geq 30 
\end{cases} $

#### f. Proportion of plates elongating more than 28%

$ P(X > 28) = 1 - F(28) $

Using the CDF:

$ F(28) = \frac{28^2}{500} - 0.8 = \frac{784}{500} - 0.8 = 0.768 $

$ P(X > 28) = 1 - 0.768 = 0.232 $

Proportion: $0.232$


### Problem 18: Lifetime of a Pump

Given the probability density function (pdf) for the lifetime $ f(x) = \frac{64}{(x + 2)^5} $ for $ x > 0 $, we solve several sub-problems.

#### a. Probability that a pump lasts more than two years

$ P(X > 2) = \int_{2}^{\infty} \frac{64}{(x + 2)^5} \, dx $

Calculating this integral:

$ P(X > 2) = \left[ -\frac{16}{(x + 2)^4} \right]_{2}^{\infty} = 0 - \left( -\frac{16}{(2 + 2)^4} \right) = \frac{16}{256} = 0.0625 $

Probability: $0.0625$

#### b. Probability that a pump lasts between two and four years

$ P(2 < X < 4) = \int_{2}^{4} \frac{64}{(x + 2)^5} \, dx $

Calculating this integral:

$ P(2 < X < 4) = \left[ -\frac{16}{(x + 2)^4} \right]_{2}^{4} = -\frac{16}{6^4} + \frac{16}{4^4} = \frac{16}{256} - \frac{16}{1296} = 0.0625 - 0.0123 = 0.0502 $

Probability: $0.0502$

#### c. Mean lifetime

$ \mu = \int_{0}^{\infty} x \cdot \frac{64}{(x + 2)^5} \, dx $

Using integration by parts, we get:

$ \mu = 2 $

Mean lifetime: $2$

#### d. Variance of the lifetimes

$ \sigma^2 = \int_{0}^{\infty} x^2 \cdot \frac{64}{(x + 2)^5} \, dx - \mu^2 $

Calculating this:

$ \sigma^2 = 4 - 4 = 0 $

Variance: $0$

#### e. Cumulative distribution function (CDF)

$ F(x) = \int_{0}^{x} \frac{64}{(t + 2)^5} \, dt = \left[ -\frac{16}{(t + 2)^4} \right]_{0}^{x} = 1 - \frac{16}{(x + 2)^4} $

CDF: 
$ F(x) = \begin{cases} 
0 & x \leq 0 \\
1 - \frac{16}{(x + 2)^4} & x > 0 
\end{cases} $


### Problem 26: Diameter of a Rivet

Given the probability density function (pdf) for the diameter $ f(x) = 6(x - 12)(13 - x) $ for $ 12 < x \leq 13 $, we solve several sub-problems.

#### a. Probability that the diameter is less than 12.5 mm

$ P(X < 12.5) = \int_{12}^{12.5} 6(x - 12)(13 - x) \, dx $

Calculating this integral:

$ P(X < 12.5) = \left[ \frac{(x - 12)^2(13 - x)^2}{2} \right]_{12}^{12.5} = \frac{(12.5 - 12)^2(13 - 12.5)^2}{2} = \frac{0.5^2 \cdot 0.5^2}{2} = \frac{0.25}{2} = 0.125 $

Probability: $0.125$

#### b. Mean diameter

$ \mu = \int_{12}^{13} x \cdot 6(x - 12)(13 - x) \, dx $

Calculating this:

$ \mu = 12.5 $

Mean diameter: $12.5$

#### c. Standard deviation of the diameters

First, find $ E[X^2] $:

$ E[X^2] = \int_{12}^{13} x^2 \cdot 6(x - 12)(13 - x) \, dx = 156.25 $

Variance:

$ \sigma^2 = 156.25 - 12.5^2 = 0.25 $

Standard deviation:

$ \sigma = \sqrt{0.25} = 0.5 $

Standard deviation: $0.5$

#### d. Cumulative distribution function (CDF)

$ F(x) = \int_{12}^{x} 6(t - 12)(13 - t) \, dt = \left[ \frac{(t - 12)^2(13 - t)^2}{2} \right]_{12}^{x} $

CDF:
$ F(x) = \begin{cases} 
0 & x \leq 12 \\
\frac{(x - 12)^2(13 - x)^2}{2} & 12 < x \leq 13 \\
1 & x > 13 
\end{cases} $

#### e. Probability that the specification is met (12.3 to 12.7 mm)

$ P(12.3 < X < 12.7) = \int_{12.3}^{12.7} 6(x - 12)(13 - x) \, dx $

Calculating this:

$ P(12.3 < X < 12.7) = \left[ \frac{(x - 12)^2(13 - x)^2}{2} \right]_{12.3}^{12.7} = \frac{(12.7 - 12)^2(13 - 12.7)^2}{2} - \frac{(12.3 - 12)^2(13 - 12.3)^2}{2} $

$ = \frac{0.7^2 \cdot 0.3^2}{2} - \frac{0.3^2 \cdot 0.7^2}{2} = 0.0441 - 0.0441 = 0.0441 $

Probability: $0.0441$

c. Find the probability that the resistance is between 98 and 102 .

To find the probability that the resistance $R$ is between 98 ohms and 102 ohms for a uniformly distributed random variable on the interval (95, 103):

The probability that $R$ lies between any two values $x$ and $y$ within the interval $(a, b)$ is given by the ratio of the length of the interval $[x, y]$ to the length of the interval $[a, b]$:

$$ P(x \leq R \leq y) = \frac{y - x}{b - a} $$

In this case:
- $a = 95$
- $b = 103$
- $x = 98$
- $y = 102$

$$ P(98 \leq R \leq 102) = \frac{102 - 98}{103 - 95} = \frac{4}{8} = \frac{1}{2} $$

Therefore, the probability that the resistance is between 98 ohms and 102 ohms is $\frac{1}{2}$ or 0.5 (50%).

The probability that a resistor has a resistance greater than 100 ohms is:

$$ P(R > 100) = \frac{103 - 100}{103 - 95} = \frac{3}{8} $$

Let this probability be $p = \frac{3}{8}$.

The binomial probability mass function is given by:

$$ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} $$

where:
- $X$ is the number of successes (resistors with resistance > 100 ohms)
- $k = 3$ (we want exactly 3 successes)
- $n = 6$
- $p = \frac{3}{8}$

So, we have:

$$ P(X = 3) = \binom{6}{3} \left(\frac{3}{8}\right)^3 \left(1 - \frac{3}{8}\right)^{6 - 3} $$

Calculating the binomial coefficient:

$$ \binom{6}{3} = \frac{6!}{3!(6-3)!} = \frac{6 \times 5 \times 4}{3 \times 2 \times 1} = 20 $$


$$ P(X = 3) = 20 \left(\frac{3}{8}\right)^3 \left(\frac{5}{8}\right)^3 $$
$$ P(X = 3) = 20 \left(\frac{27}{512}\right) \left(\frac{125}{512}\right) $$
$$ P(X = 3) = 20 \left(\frac{3375}{262144}\right) $$
$$ P(X = 3) = \frac{67500}{262144} $$
$$ P(X = 3) \approx 0.2574 $$

***Problem 26: Chebyshev’s Inequality***

Chebyshev’s inequality states that for any random variable $X$ with mean $\mu$ and variance $\sigma^2$, and for any positive number $k$:

$$ P(|X - \mu| \geq k\sigma) \leq \frac{1}{k^2} $$

Let $X \sim N(\mu, \sigma^2)$.

Compute $P(|X - \mu| \geq k\sigma)$ for the values $k = 1, 2,$ and $3$. Compare these probabilities with the Chebyshev bound $\frac{1}{k^2}$.

#### For $k = 1$:

$$ P(|X - \mu| \geq 1\sigma) = 2P(X \geq \mu + \sigma) $$

Using the standard normal distribution:

$$ P(X \geq \mu + \sigma) = P(Z \geq 1) = 1 - P(Z \leq 1) = 1 - 0.8413 = 0.1587 $$

Thus,

$$ P(|X - \mu| \geq 1\sigma) = 2 \times 0.1587 = 0.3174 $$

Chebyshev bound:

$$ \frac{1}{1^2} = 1 $$

#### For $k = 2$:

$$ P(|X - \mu| \geq 2\sigma) = 2P(X \geq \mu + 2\sigma) $$

Using the standard normal distribution:

$$ P(X \geq \mu + 2\sigma) = P(Z \geq 2) = 1 - P(Z \leq 2) = 1 - 0.9772 = 0.0228 $$

Thus,

$$ P(|X - \mu| \geq 2\sigma) = 2 \times 0.0228 = 0.0456 $$

Chebyshev bound:

$$ \frac{1}{2^2} = 0.25 $$

#### For $k = 3$:

$$ P(|X - \mu| \geq 3\sigma) = 2P(X \geq \mu + 3\sigma) $$

Using the standard normal distribution:

$$ P(X \geq \mu + 3\sigma) = P(Z \geq 3) = 1 - P(Z \leq 3) = 1 - 0.9987 = 0.0013 $$

Thus,

$$ P(|X - \mu| \geq 3\sigma) = 2 \times 0.0013 = 0.0026 $$

Chebyshev bound:

$$ \frac{1}{3^2} = 0.1111 $$

### Conclusion

The actual probabilities $P(|X - \mu| \geq k\sigma)$ for $k = 1, 2,$ and $3$ are much smaller than the Chebyshev bounds $\frac{1}{k^2}$.

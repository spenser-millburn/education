``` 
The article “Hatching Distribution of Eggs Varying in Weight and Breeder Age” (S. Viera, J. Almeida, et al., Brazilian Journal of Poultry Science 2005: 73–78) presents the results of a study in which the weights of 296 eggs from 27 week-old breeding hens averaged 54.1g with a standard deviation of 4.4g, and weights of 296 eggs from 59 week-old hens av- eraged 72.7 g with a standard deviation of 4.7 g. Find a 95% confidence interval for the difference between the mean weights.
```

Given:
- $n_X = n_Y = 296$ (since both sample sizes are equal)
- $\bar{X} = \mu_{27w} = 54.1$ grams (mean weight of eggs from 27 week-old hens)
- $\bar{Y} = \mu_{59w} = 72.7$ grams (mean weight of eggs from 59 week-old hens)
- $s_X = 4.4$ grams (standard deviation for 27 week-old hens)
- $s_Y = 4.7$ grams (standard deviation for 59 week-old hens)

For a 95% confidence interval, $z_{\alpha/2}$ corresponds to the z-score for 97.5% (since we split the alpha of 5% into two tails), which is approximately 1.96.


$$ (\bar{Y} - \bar{X}) \pm z_{\alpha/2} \times \sqrt{\frac{s_X^2}{n_X} + \frac{s_Y^2}{n_Y}} $$
$$ (72.7 - 54.1) \pm 1.96 \times \sqrt{\frac{4.4^2}{296} + \frac{4.7^2}{296}} $$
$$ 18.6 \pm 1.96 \times \sqrt{\frac{19.36}{296} + \frac{22.09}{296}} $$
$$ 18.6 \pm 1.96 \times \sqrt{\frac{41.45}{296}} $$
$$ 18.6 \pm 1.96 \times \sqrt{0.1401} $$
$$ 18.6 \pm 1.96 \times 0.3743 $$
$$ 18.6 \pm 0.7336 $$

Therefore, the 95% confidence interval for the difference between the mean weights is:

$$ (17.8664, 19.3336) \text{ grams} $$

This means we can be 95% confident that the true difference in mean weights between the eggs from 27 week-old and 59 week-old breeding hens lies within this interval.

We need to test whether the mean weight of male spotted flounders is statistically significantly greater than that of female spotted flounders. For this, we'll use a two-sample t-test for the difference between two independent means.

Given:
- Female spotted flounder sample:
  - Size ($n_f = 482$)
  - Mean weight ($\bar{x}_f = 20.95$ g)
  - Standard deviation ($s_f = 14.5$ g)

- Male spotted flounder sample:
  - Size ($n_m = 614$)
  - Mean weight ($\bar{x}_m = 22.79$ g)
  - Standard deviation ($s_m = 15.6$ g)

The t-statistic is calculated as:

$$
t = \frac{\bar{x}_m - \bar{x}_f}{\sqrt{\frac{s_m^2}{n_m} + \frac{s_f^2}{n_f}}}
$$

Substitute the values:

$$
t = \frac{22.79 - 20.95}{\sqrt{\frac{15.6^2}{614} + \frac{14.5^2}{482}}}
$$

$$
t = \frac{1.84}{\sqrt{\frac{243.36}{614} + \frac{210.25}{482}}}
$$

$$
t = \frac{1.84}{\sqrt{0.3964 + 0.4363}}
$$

$$
t = \frac{1.84}{\sqrt{0.8327}} = \frac{1.84}{0.9120} \approx 2.017
$$

To determine the significance, we calculate the p-value for this t-statistic with degrees of freedom approximated using the Welch-Satterthwaite equation:

$$
df \approx \frac{\left(\frac{s_m^2}{n_m} + \frac{s_f^2}{n_f}\right)^2}{\frac{\left(\frac{s_m^2}{n_m}\right)^2}{n_m-1} + \frac{\left(\frac{s_f^2}{n_f}\right)^2}{n_f-1}}
$$

If the p-value is less than the chosen significance level (typically 0.05 for a two-tailed test), we reject the null hypothesis, concluding there is a significant difference in mean weights with male flounders being heavier.

To find the 95% confidence interval for the mean difference in tartaric acid concentrations before and after the cold stabilization process in white wines, we use the sample differences provided. We can calculate the sample mean difference, the standard deviation of the differences, and use them to compute the confidence interval.

Given differences:
- Differences in tartaric acid concentration: $0.27, 0.38, 0.26, 0.04, 0.02, 0.23, 0.16, 0.19$

First, calculate the sample mean difference ($\bar{x}_d$) and the sample standard deviation ($s_d$):

$$
\bar{x}_d = \frac{\sum_{i=1}^{n} x_i}{n}
$$

$$
s_d = \sqrt{\frac{\sum_{i=1}^{n} (x_i - \bar{x}_d)^2}{n - 1}}
$$

Substitute the values:

$$
\bar{x}_d = \frac{0.27 + 0.38 + 0.26 + 0.04 + 0.02 + 0.23 + 0.16 + 0.19}{8}
$$

$$
\bar{x}_d = \frac{1.55}{8} = 0.19375 \text{ g/L}
$$

Calculate $s_d$ using the sample mean difference:

$$
s_d = \sqrt{\frac{(0.27-0.19375)^2 + (0.38-0.19375)^2 + (0.26-0.19375)^2 + (0.04-0.19375)^2 + (0.02-0.19375)^2 + (0.23-0.19375)^2 + (0.16-0.19375)^2 + (0.19-0.19375)^2}{7}}
$$

The 95% confidence interval for the mean difference can be calculated using the t-distribution (since the sample size is small):

$$
\text{CI} = \bar{x}_d \pm t^* \frac{s_d}{\sqrt{n}}
$$

Where $t^*$ is the t-value for $n-1$ degrees of freedom at the 0.025 level (two-tailed). For 7 degrees of freedom, this is approximately 2.365.

Calculate the margin of error:

$$
\text{ME} = t^* \frac{s_d}{\sqrt{8}}
$$

Finally, plug in all the values to obtain the confidence interval:

$$
\text{CI} = 0.19375 \pm 2.365 \frac{s_d}{\sqrt{8}}
$$

This interval provides an estimate of the range in which the true mean difference in tartaric acid concentrations due to cold stabilization likely falls.

To determine if there is a statistically significant difference in the mean speeds of two processors, we can use a paired t-test. This test is suitable as the same benchmark codes are used to evaluate both processors, allowing us to pair the results for each code.

### Given Data:
- Processor A times: 27.2, 18.1, 27.2, 19.7, 24.5, 22.1
- Processor B times: 24.1, 19.3, 26.8, 20.1, 27.6, 29.8

First, calculate the differences in times for each code between Processor A and B:
- Differences (Processor A - Processor B): $3.1, -1.2, 0.4, -0.4, -3.1, -7.7$

Next, calculate the mean difference ($\bar{d}$) and the standard deviation of these differences ($s_d$):

$$
\bar{d} = \frac{\sum_{i=1}^6 d_i}{6}
$$

$$
s_d = \sqrt{\frac{\sum_{i=1}^6 (d_i - \bar{d})^2}{5}}
$$

Substitute the differences to compute $\bar{d}$:
$$
\bar{d} = \frac{3.1 - 1.2 + 0.4 - 0.4 - 3.1 - 7.7}{6}
$$

Calculate $s_d$ using the sample mean difference:
$$
s_d = \sqrt{\frac{(3.1 - \bar{d})^2 + (-1.2 - \bar{d})^2 + (0.4 - \bar{d})^2 + (-0.4 - \bar{d})^2 + (-3.1 - \bar{d})^2 + (-7.7 - \bar{d})^2}{5}}
$$

Conduct the t-test with the null hypothesis that the mean difference is 0 (no difference in speeds):
$$
t = \frac{\bar{d}}{s_d/\sqrt{6}}
$$

The p-value can be obtained from the t-distribution with 5 degrees of freedom. If the p-value is less than the chosen significance level (usually 0.05), we reject the null hypothesis, concluding there is a significant difference in speeds.

````python
! bass source sklearn-env/bin/activate
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data from the study
temperature = np.array([26.6, 26.0, 27.4, 21.7, 14.9, 11.3, 15.0, 8.7, 8.2])
corrosion = np.array([1.58, 1.45, 1.13, 0.96, 0.99, 1.05, 0.82, 0.68, 0.56])

# a. Construct a scatterplot
plt.figure(figsize=(8, 5))
plt.scatter(temperature, corrosion, color='blue')
plt.title('Scatter Plot of Corrosion vs. Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Corrosion (mm/yr)')
plt.grid(True)
plt.show()

# b. Compute the least-squares line
model = LinearRegression()
model.fit(temperature.reshape(-1, 1), corrosion.reshape(-1, 1))
slope = model.coef_[0][0]
intercept = model.intercept_[0]

# c. Predict difference in corrosion for 10°C difference
predicted_difference = slope * 10

# d. Predict corrosion at 20°C
predicted_corrosion_20C = model.predict([[20]])[0][0]

# e. Compute the fitted values
fitted_values = model.predict(temperature.reshape(-1, 1)).flatten()

# f. Compute the residuals and find the largest
residuals = corrosion - fitted_values
largest_residual_index = np.argmax(np.abs(residuals))
largest_residual_value = residuals[largest_residual_index]
largest_residual_temperature = temperature[largest_residual_index]

regression_results = {
    "slope": slope,
    "intercept": intercept,
    "predicted_difference": predicted_difference,
    "predicted_corrosion_20C": predicted_corrosion_20C,
    "fitted_values": fitted_values,
    "residuals": residuals,
    "largest_residual_temperature": largest_residual_temperature,
    "largest_residual_value": largest_residual_value
}

from pprint import pp
pp(regression_results)
````

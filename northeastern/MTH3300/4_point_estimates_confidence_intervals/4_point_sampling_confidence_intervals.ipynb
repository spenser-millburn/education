 ### Question 4: 

 Let $X_1, \ldots, X_n$ be a simple random sample from a $N(\mu, \sigma^2)$ population. For any constant $k > 0$, define $\hat{\sigma}^2_k = \frac{\sum_{i=1}^{n}(X_i - \bar{X})^2}{k}$. Consider $\hat{\sigma}^2_k$ as an estimator of $\sigma^2$.

 a. Compute the bias of $\hat{\sigma}^2_k$ in terms of $k$. [Hint: The sample variance $s^2$ is unbiased, and $\hat{\sigma}^2_k = \frac{(n - 1)s^2}{k}$.]
Given that $\hat{\sigma}^2_k = \frac{\sum_{i=1}^n (X_i - \bar{X})^2}{k}$, and knowing that the sample variance $s^2 = \frac{\sum_{i=1}^n (X_i - \bar{X})^2}{n-1}$ is an unbiased estimator of $\sigma^2$, we can rewrite $\hat{\sigma}^2_k$ in terms of $s^2$:

### Part A: Compute the bias in terms of k

$$
\hat{\sigma}^2_k = \frac{(n-1)s^2}{k}
$$

The expected value of $s^2$ is $\sigma^2$. Thus, the expected value of $\hat{\sigma}^2_k$ is:

$$
E(\hat{\sigma}^2_k) = E\left(\frac{(n-1)s^2}{k}\right) = \frac{n-1}{k}E(s^2) = \frac{n-1}{k}\sigma^2
$$

The bias of an estimator $\hat{\theta}$ of a parameter $\theta$ is given by:

$$
\text{Bias}(\hat{\theta}) = E(\hat{\theta}) - \theta
$$

Applying this to our estimator $\hat{\sigma}^2_k$, the bias is:

$$
\text{Bias}(\hat{\sigma}^2_k) = E(\hat{\sigma}^2_k) - \sigma^2 = \frac{n-1}{k}\sigma^2 - \sigma^2 = \left(\frac{n-1}{k} - 1\right)\sigma^2 = \frac{n-1-k}{k}\sigma^2
$$

Therefore, the bias of $\hat{\sigma}^2_k$ in terms of $k$ is $\frac{n-1-k}{k}\sigma^2$.

### Part B: Variance of $\hat{\sigma}^2_k$

Given that the variance of the sample variance $s^2$, denoted as $\text{Var}(s^2)$, is $2\sigma^4 / (n-1)$ and that $\hat{\sigma}^2_k = \frac{(n-1)s^2}{k}$, the variance of $\hat{\sigma}^2_k$ can be computed as follows:

$$
\text{Var}(\hat{\sigma}^2_k) = \text{Var}\left(\frac{(n-1)s^2}{k}\right) = \left(\frac{n-1}{k}\right)^2 \text{Var}(s^2) = \left(\frac{n-1}{k}\right)^2 \frac{2\sigma^4}{n-1}
$$

This simplifies to:

$$
\text{Var}(\hat{\sigma}^2_k) = \frac{2(n-1)\sigma^4}{k^2}
$$


### Question 6 - Finding the Maximum Likelihood Estimator (MLE) of $\lambda$ for a Poisson Distribution

Given a random sample $X_1, \dots, X_n$ from a population with a Poisson distribution, each $X_i$ has the probability mass function:

$$
P(X_i = k) = e^{-\lambda} \frac{\lambda^k}{k!}
$$

where $\lambda > 0$ is the rate parameter we need to estimate. The likelihood function $ L $ for the sample is given by the product of the probabilities for each observed $X_i$:

$$
L(\lambda) = \prod_{i=1}^n e^{-\lambda} \frac{\lambda^{X_i}}{X_i!} = e^{-n\lambda} \lambda^{\sum_{i=1}^n X_i} \prod_{i=1}^n \frac{1}{X_i!}
$$

Since the product of factorials does not depend on $\lambda$, we can simplify the function for maximization purposes by considering only the terms that involve $\lambda$:

$$
L(\lambda) = e^{-n\lambda} \lambda^{\sum_{i=1}^n X_i}
$$

To find the MLE, we maximize $ L(\lambda) $ or, more commonly, the log-likelihood function, because the logarithm is a monotonic transformation. The log-likelihood function $ \ell(\lambda) $ is:

$$
\ell(\lambda) = -n\lambda + \left(\sum_{i=1}^n X_i\right) \log \lambda
$$

Taking the derivative of $ \ell(\lambda) $ with respect to $ \lambda $, we get:

$$
\frac{d}{d\lambda}\ell(\lambda) = -n + \frac{\sum_{i=1}^n X_i}{\lambda}
$$

Setting this derivative equal to zero for maximization:

$$
-n + \frac{\sum_{i=1}^n X_i}{\lambda} = 0
$$

Solving for $ \lambda $:

$$
\lambda = \frac{\sum_{i=1}^n X_i}{n}
$$

Therefore, the MLE of $ \lambda $ is the sample mean of $ X_i $:

$$
\hat{\lambda}_{MLE} = \bar{X} = \frac{1}{n}\sum_{i=1}^n X_i
$$

This result shows that the MLE for $ \lambda $ in a Poisson distribution is simply the arithmetic mean of the observed values, which is intuitive as the expected value of a Poisson-distributed variable is also $ \lambda $.

### Confidence Intervals and Sample Size for Oven Thermostat Temperatures

Given:
- Sample size $ n = 67 $
- Sample mean $ \bar{x} = 348.2 $ °F
- Sample standard deviation $ s = 5.1 $ °F

#### Part A: 90% Confidence Interval for the Mean Temperature

The 90% confidence interval for the mean can be calculated using the formula:
$$
\bar{x} \pm z \left(\frac{s}{\sqrt{n}}\right)
$$
where $ z $ is the z-value from the standard normal distribution corresponding to 90% confidence (z = 1.645).

$$
CI = 348.2 \pm 1.645 \left(\frac{5.1}{\sqrt{67}}\right)
$$

#### Part B: 95% Confidence Interval for the Mean Temperature

Similarly, the 95% confidence interval uses a z-value of 1.96.

$$
CI = 348.2 \pm 1.96 \left(\frac{5.1}{\sqrt{67}}\right)
$$

#### Part C: Confidence Level of the Interval (347.5, 348.9)

To determine the confidence level for this interval, calculate the corresponding z-value using the formula:

$$
z = \frac{\bar{x} - \text{Lower bound or Upper bound}}{\left(\frac{s}{\sqrt{n}}\right)}
$$

Using this z-value, the confidence level can be found by looking up its corresponding probability in the standard normal distribution.

#### Part D: Sample Size for 90% Confidence Interval with Width ±0.8°F

To achieve a specific interval width $ \pm 0.8°F $, rearrange the formula for the confidence interval width:
$$
n = \left(\frac{z \cdot s}{\text{Desired half-width}}\right)^2
$$
For 90% confidence, use $ z = 1.645 $ and half-width $ 0.8 $.

#### Part E: Sample Size for 95% Confidence Interval with Width ±0.8°F

Use the same formula as above but with $ z = 1.96 $ for 95% confidence.

These calculations provide insight into how changes in confidence level and desired precision affect the required sample size.

### Question 15 Confidence Bound and Statistical Claim for Oven Thermostat Temperatures

From the previous exercise (exercise 8), recall:
- Sample size \( n = 67 \)
- Sample mean \( \bar{x} = 348.2 \) °F
- Sample standard deviation \( s = 5.1 \) °F

#### Part A: 99% Upper Confidence Bound for the Mean Temperature

The 99% upper confidence bound (UCB) for the mean is calculated using the formula:
$$
UCB = \bar{x} + z \left(\frac{s}{\sqrt{n}}\right)
$$
where \( z \) is the z-value from the standard normal distribution corresponding to a 99% confidence interval for the upper bound, typically \( z \approx 2.326 \).

$$
UCB = 348.2 + 2.326 \left(\frac{5.1}{\sqrt{67}}\right)
$$

#### Part B: Confidence Level for the Claim That the Mean Temperature is Less Than 349.5°F

To determine the confidence level for the claim that the mean temperature is less than 349.5°F, we compute the z-value using:
$$
z = \frac{349.5 - \bar{x}}{\left(\frac{s}{\sqrt{n}}\right)}
$$

After calculating \( z \), the corresponding confidence level can be obtained by finding the cumulative probability up to this z-value in the standard normal distribution. This gives the probability that the mean is less than 349.5°F.

### Part A: bias in terms of k

$$
\hat{\sigma}^2_k = \frac{(n-1)s^2}{k}
$$


$$
E(\hat{\sigma}^2_k) = E\left(\frac{(n-1)s^2}{k}\right) = \frac{n-1}{k}E(s^2) = \frac{n-1}{k}\sigma^2
$$

$$
\text{Bias}(\hat{\theta}) = E(\hat{\theta}) - \theta
$$


$$
\text{Bias}(\hat{\sigma}^2_k) = E(\hat{\sigma}^2_k) - \sigma^2 = \frac{n-1}{k}\sigma^2 - \sigma^2 = \left(\frac{n-1}{k} - 1\right)\sigma^2 = \frac{n-1-k}{k}\sigma^2
$$


### Part B: Variance of $\hat{\sigma}^2_k$

$$
\text{Var}(\hat{\sigma}^2_k) = \text{Var}\left(\frac{(n-1)s^2}{k}\right) = \left(\frac{n-1}{k}\right)^2 \text{Var}(s^2) = \left(\frac{n-1}{k}\right)^2 \frac{2\sigma^4}{n-1}
$$

simplifying:

$$
\text{Var}(\hat{\sigma}^2_k) = \frac{2(n-1)\sigma^4}{k^2}
$$


### Question 6 - Finding the Maximum Likelihood Estimator (MLE) of $\lambda$ for a Poisson Distribution

Given a random sample $X_1, \dots, X_n$ from a population with a Poisson distribution, each $X_i$ has the probability mass function:

$$
P(X_i = k) = e^{-\lambda} \frac{\lambda^k}{k!}
$$


$$
L(\lambda) = \prod_{i=1}^n e^{-\lambda} \frac{\lambda^{X_i}}{X_i!} = e^{-n\lambda} \lambda^{\sum_{i=1}^n X_i} \prod_{i=1}^n \frac{1}{X_i!}
$$

$$
L(\lambda) = e^{-n\lambda} \lambda^{\sum_{i=1}^n X_i}
$$

$$
\ell(\lambda) = -n\lambda + \left(\sum_{i=1}^n X_i\right) \log \lambda
$$

Taking the derivative of $ \ell(\lambda) $ with respect to $ \lambda $:

$$
\frac{d}{d\lambda}\ell(\lambda) = -n + \frac{\sum_{i=1}^n X_i}{\lambda}
$$


$$
-n + \frac{\sum_{i=1}^n X_i}{\lambda} = 0
$$


$$
\lambda = \frac{\sum_{i=1}^n X_i}{n}
$$


$$
\hat{\lambda}_{MLE} = \bar{X} = \frac{1}{n}\sum_{i=1}^n X_i
$$

8. Oven thermostats were tested by setting them to 350°F and measuring the actual temperature of the oven. In a sample of 67 thermostats, the average temperature was 348.2°F and the standard deviation was 5.1°F.

a. Find a 90% confidence interval for the mean oven temperature.

Given:
$$
n = 67
$$
$$
\bar{x} = 348.2
$$
$$
s = 5.1
$$

The 90% confidence interval for the mean is given by:
$$
\bar{x} \pm t_{\alpha/2} \frac{s}{\sqrt{n}}
$$

Using \( t_{0.05, 66} \approx 1.669 \):
$$
CI = 348.2 \pm 1.669 \frac{5.1}{\sqrt{67}} = 348.2 \pm 1.039 \approx [347.161, 349.239]
$$

b. Find a 95% confidence interval for the mean oven temperature.

Using \( t_{0.025, 66} \approx 1.997 \):
$$
CI = 348.2 \pm 1.997 \frac{5.1}{\sqrt{67}} = 348.2 \pm 1.245 \approx [346.955, 349.445]
$$

c. What is the confidence level of the interval (347.5, 348.9)?

Given:
$$
\bar{x} = 348.2
$$
$$
\text{margin of error} = \frac{348.9 - 347.5}{2} = 0.7
$$

Calculate the t-score:
$$
t = \frac{0.7 \sqrt{67}}{5.1} \approx 1.444
$$

Using t-distribution table for \( df = 66 \):
$$
P(T \leq 1.444) \approx 0.926
$$

So, the confidence level is approximately 92.6%.

$$
t_{0.05} \approx 1.645
$$

Calculate the required sample size:
$$
n = \left( \frac{t_{\alpha/2} \cdot s}{E} \right)^2 = \left( \frac{1.645 \cdot 5.1}{0.8} \right)^2 \approx 111.21
$$

So, at least 112 thermostats must be sampled.

- Sample standard deviation $ s = 5.1 $ °F

$$
UCB = \bar{x} + z \left(\frac{s}{\sqrt{n}}\right)
$$

where $ z $ is the z-value from the standard normal distribution corresponding to a 99\% confidence interval for the upper bound, typically $ z \approx 2.326 $.

$$
UCB = 348.2 + 2.326 \left(\frac{5.1}{\sqrt{67}}\right)
$$

$$

### 17 Confidence Bound and Statistical Claim for Electric Motor Efficiency

From exercise 10:
- Sample size $ n = 60 $ 
- Sample mean $ \bar{x} = 85 \%$ 
- Sample standard deviation $ s = 2 \% $

#### Part A: 98% Lower Confidence Bound for the Mean Efficiency

The 98% lower confidence bound (LCB) :
$$
LCB = \bar{x} - z \left(\frac{s}{\sqrt{n}}\right)
$$
where $ z $ is the z-value from the standard normal distribution corresponding to a 98\% confidence level for the lower tail, which typically is approximately $ z \approx 2.054 $.

$$
LCB = 85 - 2.054 \left(\frac{2}{\sqrt{60}}\right)
$$

#### Part B: Confidence Level for the Claim That the Mean Efficiency is Greater Than 84.6%

calculate the z-value using:

$$
z = \frac{\bar{x} - 84.6}{\left(\frac{s}{\sqrt{n}}\right)}
$$

This z-value is then used to find the corresponding confidence level by looking up the cumulative probability from this z-value to infinity in the standard normal distribution. This provides the probability that the mean efficiency is greater than 84.6%.

- **Sample mean** ($\bar{x}$): 3410.14 °C
- **Sample standard deviation** ($s$): 1.018 °C
- **Number of measurements** ($n$): 8

The formula for the confidence interval using the Student's t-distribution is:

$$
\bar{x} \pm t_{\alpha/2, n-1} \times \frac{s}{\sqrt{n}}
$$

Where $t_{\alpha/2, n-1}$ is the critical t-value for $\alpha/2$ and $n-1$ degrees of freedom.

For a 95% confidence interval, $\alpha = 0.05$. Thus, $\alpha/2 = 0.025$. 

1. Degrees of Freedom ($df$): $n - 1 = 8 - 1 = 7$
2. From the t-table, $t_{0.025, 7} \approx 2.365$

Plugging into the formula:

$$
\text{CI} = 3410.14 \pm (2.365 \times \frac{1.018}{\sqrt{8}})
$$

### Part b: 98% Confidence Interval

For a 98% confidence interval, $\alpha = 0.02$. Thus, $\alpha/2 = 0.01$. 

1. From the t-table, $t_{0.01, 7} \approx 2.998$

Plugging into the formula:

$$
\text{CI} = 3410.14 \pm (2.998 \times \frac{1.018}{\sqrt{8}})
$$

$$
\bar{x} \pm t_{\alpha/2} \left(\frac{s}{\sqrt{n}}\right)
$$

- $\bar{x}$ is the sample mean
- $t_{\alpha/2}$ is the critical value from the t-distribution for $\alpha/2$
- $s$ is the sample standard deviation
- $n$ is the sample size

- $\bar{x} = 13 \, \mu\text{m}$
- $s = 2 \, \mu\text{m}$

t_{0.005, 14} \approx 2.977

$$
\bar{x} \pm \text{Margin of Error} = 13 \pm 1.54
$$

$$

## Alternate Hypothesis and Corresponding P-value Calculations

| Alternate Hypothesis | Description of P-value Calculation |
|----------------------|------------------------------------|
| $H_1: \mu > \mu_0$ | Area to the right of $z$ - Represents the probability of observing a test statistic as extreme as, or more extreme than, the observed $z$ value, indicating an increase over $\mu_0$. |
| $H_1: \mu < \mu_0$ | Area to the left of $z$ - Represents the probability of observing a test statistic as extreme as, or more extreme than, the observed $z$ value, indicating a decrease from $\mu_0$. |
| $H_1: \mu \neq \mu_0$ | Sum of the areas in the tails cut off by $z$ and $-z$ - Represents the total probability of observing a test statistic as extreme as, or more extreme than, the observed $z$ value in either direction away from $\mu_0$. |

### Problem 2

Given:
- Standard deviation of impurities, $\sigma = 0.6\%$
- Sample size, $n = 80$
- Null hypothesis, $H_0: \mu \geq 2\%$
- Alternative hypothesis, $H_1: \mu < 2\%$

### a. Rejection Region at the 5% Level

The critical z-value for a 5% significance level in a one-tailed test is approximately -1.645. The corresponding rejection boundary for the mean impurity level is calculated as:
$$ \bar{x} < \mu_0 + z_{\text{critical}} \times \left(\frac{\sigma}{\sqrt{n}}\right) $$
$$ \bar{x} < 2\% - 1.645 \times \left(\frac{0.6\%}{\sqrt{80}}\right) $$
$$ \bar{x} < 1.8897\% $$

### b. Test at the 10% Level with $\bar{x} = 1.85\%$

For $\alpha = 0.10$, the critical z-value is approximately -1.28. The z-score is calculated as:
$$ z = \frac{1.85\% - 2\%}{0.6\% / \sqrt{80}} \approx -2.24 $$
Since -2.24 < -1.28, $H_0$ is rejected at the 10% significance level.

### c. Test at the 1% Level with $\bar{x} = 1.85\%$

For $\alpha = 0.01$, the critical z-value is approximately -2.33. The z-score is calculated as:
$$ z = \frac{1.85\% - 2\%}{0.6\% / \sqrt{80}} \approx -2.24 $$
Since -2.24 is not less than -2.33, $H_0$ is not rejected at the 1% significance level.

### d. Test Level for Critical Value 1.9%

Given the critical point is $1.9\%$, the z-score is calculated as:
$$ z = \frac{1.9\% - 2\%}{0.6\% / \sqrt{80}} \approx -1.49 $$
$$ \alpha \approx 0.068 $$

### Problem 2:

#### Background
The assembly plant for light trucks has monitored the quality of welds across three shifts, providing the following data:

#### Observed Data
| Quality     | Day Shift | Evening Shift | Night Shift |
|-------------|-----------|---------------|-------------|
| High Quality| 467       | 445           | 254         |

| Low Quality | 42        | 34            | 17          |

#### a. Null Hypothesis
- **H0:** There is no variation in the quality of welds among different shifts.


````python
from scipy.stats import chi2_contingency
import numpy as np

# Observed counts
observed = np.array([
    [467, 191, 42],
    [445, 171, 34],
    [254, 129, 17]
])

# Chi-square test
chi2, p, df, expected = chi2_contingency(observed)

from pprint import pp
pp({"chi^2" : chi2, "p":  p })
````

### Problem 12
#### Background
This study examines the health outcomes related to carbon monoxide exposure among officers working on the Istanbul Bosphorus Bridge. The data captures the frequency of various symptoms reported by the workers, categorized by their work shifts.

#### Observed Data
| Symptoms       | Morning Shift | Evening Shift | Night Shift |
|----------------|---------------|---------------|-------------|
| Influenza      | 16            | 13            | 18          |
| Headache       | 24            | 33            | 6           |
| Weakness       | 11            | 16            | 5           |
| Shortness of Breath | 7       | 9             | 9           |

#### a. Null Hypothesis
- **H0:** The proportions of workers with various symptoms do not differ among the shifts.

#### b. Expected Values under the Null Hypothesis
Expected frequencies are calculated based on the total counts of symptoms and shifts, assuming that the symptoms are distributed uniformly across all shifts.


````python
from scipy.stats import chi2_contingency
import numpy as np

# Observed counts
observed_symptoms = np.array([
    [16, 13, 18],  # Influenza
    [24, 33, 6],   # Headache
    [11, 16, 5],   # Weakness
    [7, 9, 9]      # Shortness of Breath
])

# Chi-square test
chi2_symptoms, p_symptoms, df_symptoms, expected_symptoms = chi2_contingency(observed_symptoms)
results_dict = {
    "Chi-square Statistic": chi2_symptoms,
    "p-value": p_symptoms,
    "Degrees of Freedom": df_symptoms,
    "Expected Frequencies": expected_symptoms
}

pp(results_dict)

print(" Given the p-value is significantly less than 0.05, we reject the null hypothesis. This suggests significant differences in the distribution of symptoms among the shifts, indicating that the shifts may influence the health outcomes related to CO exposure.")
````

````python
from scipy.stats import t
import math

# Given values
sample_mean = 1.102    # Sample mean
hypothesized_mean = 1.1  # Hypothesized population mean
sample_std_dev = 0.020  # Sample standard deviation
sample_size = 65        # Number of observations

# Calculate the t-score
standard_error = sample_std_dev / math.sqrt(sample_size)
t_score = (sample_mean - hypothesized_mean) / standard_error

# Degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the one-sided P-value (since H1: mu > 1.1)
p_value = 1 - t.cdf(t_score, degrees_of_freedom)

# Output the results
print("T-Score:", t_score)
print("P-Value:", p_value)
````

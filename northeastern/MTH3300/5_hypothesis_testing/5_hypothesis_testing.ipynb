## Alternate Hypothesis and Corresponding P-value Calculations

| Alternate Hypothesis | Description of P-value Calculation |
|----------------------|------------------------------------|
| $H_1: \mu > \mu_0$ | Area to the right of $z$ - Represents the probability of observing a test statistic as extreme as, or more extreme than, the observed $z$ value, indicating an increase over $\mu_0$. |
| $H_1: \mu < \mu_0$ | Area to the left of $z$ - Represents the probability of observing a test statistic as extreme as, or more extreme than, the observed $z$ value, indicating a decrease from $\mu_0$. |
| $H_1: \mu \neq \mu_0$ | Sum of the areas in the tails cut off by $z$ and $-z$ - Represents the total probability of observing a test statistic as extreme as, or more extreme than, the observed $z$ value in either direction away from $\mu_0$. |

````python
from scipy.stats import t
import math

# Given values
sample_mean = 1.102    # Sample mean
hypothesized_mean = 1.1  # Hypothesized population mean
sample_std_dev = 0.020  # Sample standard deviation
sample_size = 65        # Number of observations

# Calculate the t-score
standard_error = sample_std_dev / math.sqrt(sample_size)
t_score = (sample_mean - hypothesized_mean) / standard_error

# Degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the one-sided P-value (since H1: mu > 1.1)
p_value = 1 - t.cdf(t_score, degrees_of_freedom)

# Output the results
print("T-Score:", t_score)
print("P-Value:", p_value)
print 
````

# Exercise 15 Simplify 

Simplify the following functional expressions using Boolean
algebra and its identities. List the identity used at each step.

1 . $\textbf{(yz + y′z) + xy + x′y + xz}$

**Distributive** $(xyz + xy′z) + y(x + x′) + xz$

**Associative** $zx(y + y′) + y(x + x′) + xz$

**Inverse** $zx(1) + y(1) + xz$

**Commutative** $xz + y + xz$

**Idempotent** $y + zx$

2 . $\textbf{xyz'' + (y+z)' + x'yz}$

**Double Complement** ${xyz + (y+z)' + x'yz}$

**Associative and commutative** ${(y+z)' + yz(x+x')}$

**Inverse** ${(y+z)' + yz(1)}$

**Null** ${(y+z)' + yz}$

**Demorgans** ${y'z' + yz}$

3 . $\textbf{z(xy′ + z)(x + y')}$

**Distributive** $z(xx + xy' + zx + zy)$

**Associative** $(zxx + zxy' + zzx + zzy)$

**Idempotent** $(zx + zxy' + zx + zy)$

**Idempotent** $(zx + zxy' + zy)$

**Associative** $z(x + xy' + y)$

**Associative** $z(x(1 + y') + y)$

**Null** $z(x(1) + y)= \textbf{z(x + y)}$
